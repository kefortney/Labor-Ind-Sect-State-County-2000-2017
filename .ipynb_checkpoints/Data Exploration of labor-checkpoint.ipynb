```python
import pandas as pd
import numpy as np
# Visualization
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
%matplotlib inline
```

```python
# Load cumilative datasets created in step 1
cumllaborcnty = pd.read_csv('cumllaborcnty.csv')
cumllaborvt = pd.read_csv('cumllaborvt.csv')
```

```python
#Take all the months and stack vertically them instead of having them horizontal
Jan = cumllaborcnty.loc[:,['Year','County','Type','Jan']]
Jan['Month'] = 'January'
Jan.columns = ['Year','County','Type','Count','Month']

Feb = cumllaborcnty.loc[:,['Year','County','Type','Feb']]
Feb['Month'] = 'February'
Feb.columns = ['Year','County','Type','Count','Month']

Mar = cumllaborcnty.loc[:,['Year','County','Type','Mar']]
Mar['Month'] = 'March'
Mar.columns = ['Year','County','Type','Count','Month']

Apr = cumllaborcnty.loc[:,['Year','County','Type','Apr']]
Apr['Month'] = 'April'
Apr.columns = ['Year','County','Type','Count','Month']

May = cumllaborcnty.loc[:,['Year','County','Type','May']]
May['Month'] = 'May'
May.columns = ['Year','County','Type','Count','Month']

Jun = cumllaborcnty.loc[:,['Year','County','Type','Jun']]
Jun['Month'] = 'June'
Jun.columns = ['Year','County','Type','Count','Month']

Jul = cumllaborcnty.loc[:,['Year','County','Type','Jul']]
Jul['Month'] = 'July'
Jul.columns = ['Year','County','Type','Count','Month']

Aug = cumllaborcnty.loc[:,['Year','County','Type','Aug']]
Aug['Month'] = 'August'
Aug.columns = ['Year','County','Type','Count','Month']

Sep = cumllaborcnty.loc[:,['Year','County','Type','Sep']]
Sep['Month'] = 'September'
Sep.columns = ['Year','County','Type','Count','Month']

Oct = cumllaborcnty.loc[:,['Year','County','Type','Oct']]
Oct['Month'] = 'October'
Oct.columns = ['Year','County','Type','Count','Month']

Nov = cumllaborcnty.loc[:,['Year','County','Type','Nov']]
Nov['Month'] = 'November'
Nov.columns = ['Year','County','Type','Count','Month']

Dec = cumllaborcnty.loc[:,['Year','County','Type','Dec']]
Dec['Month'] = 'December'
Dec.columns = ['Year','County','Type','Count','Month']

# Now put them all together 
monthsdf = [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]
months = pd.concat(monthsdf, ignore_index=True)
```

```python
# There are missing values or occasions where the occurance is too small to show, each we are replacing with NaNs
months = months.replace('(c)', np.nan)
months = months.replace('-', np.nan)
months['Count'] = pd.to_numeric(months['Count'])
```

```python
# Clean up the column order and then sort everythin
months = months[['Year','Month','County','Type','Count']]
months.sort_values(by=['County','Year','Month'])
months.shape
```

## Exploring the Data

At this point what we have is month over month employment counts for 17 years by industry at various levels. First we are going to tackle the supersectors and see what they look like in box plots. Box plots are a standardized way of displaying the distribution of data based on the five number summary: 
* minimum (bottom T of the line)
* first quartile (bottom of the box)
* median (horizontal line bisecting the box)
* third quartile (top of the box)
* maximum. (top T of the line

```python
natres = months[months['Type'] == 'Natural Resources and Mining supersector']

data = natres
sns.boxplot(x='Month', y='Count',  data=data)
plt.xticks(rotation=90)
```

```python
data = natres
sns.pointplot(x='Year', y='Count', hue='County',  data=data)
plt.xticks(rotation=45)
```
